# Goal
The goal of this notebook is to test the neural network implemented in solution.py on the SVHN (The Street View House Numbers) dataset which is as indicated a set of images of house numbers thus it is a classification task. We also try some methods to improve the accuracy of the model.

# Prepare the data and the model

```python
# Import libraries
from solution import NN
import matplotlib.pyplot as plt
```

```python
# Initialize the neural network
nn = NN(datapath='../data/svhn.pkl',lr=0.03,batch_size=100, seed=0)
```

# Train the model and plot the curves

```python
# Train the model
train_logs = nn.train_loop(n_epochs=25)
```

```python
# Plot the accuracy curves
plt.plot(train_logs['train_accuracy'], label='train accuracy')
plt.plot(train_logs['validation_accuracy'],  label='validation accuracy')
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.ylim((0.2, 1))
plt.legend()
plt.show()
```

```python
# Plot the loss curves
plt.plot(train_logs['train_loss'], label='train loss')
plt.plot(train_logs['validation_loss'], label='validation loss')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.ylim((0, 2.3))
plt.legend()
plt.show()
```

# Normalization
One popular regularization technique is called normalization, the one we will use here is standard normalization i.e centering and scaling the data before training

```python
# Load and normalize the data then train
nn_norm = NN(datapath='../data/svhn.pkl',lr=0.03,batch_size=100, seed=0, normalization=True)
train_logs = nn_norm.train_loop(25)
```

```python
# Plot the accuracy curves
plt.plot(train_logs['train_accuracy'], label='train accuracy')
plt.plot(train_logs['validation_accuracy'],  label='validation accuracy')
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.ylim((0.2, 1))
plt.legend()
plt.show()
```

```python
# Plot the loss curves
plt.plot(train_logs['train_loss'], label='train_loss')
plt.plot(train_logs['validation_loss'], label='validation_loss')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.ylim((0, 2.3))
plt.legend()
plt.show()
```

# Trading width for depth
A well known concept is the efficiency of depth over width, essentially a deep neural network will perform better than a shallow one with the same number of parameters, let's verify this

```python
# Prepare and train the model
nn_deep = NN(hidden_dims=(512, 120, 120, 120, 120, 120, 120), datapath='../data/svhn.pkl',lr=0.03,batch_size=100, seed=0)
train_logs = nn_deep.train_loop(25)
```

```python
# Plot the accuracy curves
plt.plot(train_logs['train_accuracy'], label='train accuracy')
plt.plot(train_logs['validation_accuracy'],  label='validation accuracy')
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.ylim((0.2, 1))
plt.legend()
plt.show()
```

```python
# Plot the loss curves
plt.plot(train_logs['train_loss'], label='train_loss')
plt.plot(train_logs['validation_loss'], label='validation_loss')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.ylim((0, 2.3))
plt.legend()
plt.show()
```
